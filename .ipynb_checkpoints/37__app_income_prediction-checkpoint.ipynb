# Aplikacja Przewidywanie Przychodu powyżej 50K

## Specyfikacja / Specs

* Aplikacja ma za zadanie przewidzieć czy dana osoba zarabia powyżej 50K rocznie.
* Użytkownik wprowadza dane osobowe, a aplikacja zwraca predykcję.
* Aplikacja prezentuje najważniejsze cechy, które wpływają na predykcję.

## Plan / Taski
* v1 - możliwość wpisania danych 
* v2 - predykcja czy przychód powyżej 50K
* v3 - prezentacja najważniejszych cech

## Zależności
* `streamlit` - do budowy interfejsu użytkownika.
* `pycaret` - do budowy modelu ML.

## Instalujemy wymagane biblioteki (conda wymagana)

- otwórz terminal
- uruchom `conda activate od_zera_do_ai`
- uruchom `conda config --append channels conda-forge`
- uruchom `conda install -y streamlit`
- uruchom `conda install -y 'pycaret[full]'`
